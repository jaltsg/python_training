# Types of Data Fields

There are 3 main types of data types. Knowing what type of data points they are is the first step into knowing what kind of analyses you can do on the dataset.

<b>Identifiers</b>:
Data fields that are used to uniquely identify <b>entities of interest</b> e.g. IC number, company registration number, purchase order number

<b>Categorical</b>:
Data fields that allow the data set to be <b>grouped</b> or clustered or organized. E.g. gender, occupation,  department

<b>Measures</b>:
Data fields that hold numbers that quantify the <b>value</b> of an entity. E.g. Age, price, quantity
