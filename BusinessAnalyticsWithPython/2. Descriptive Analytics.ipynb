# Anomaly Detection using Descriptive Statistics

```python
# some data points are best plotted out as scattered points across a scale in order visualize the distribution
# a box plot a is a special type of graph that 
salesDF["Shipping Cost"].plot(kind="box")
```

The anatomy of a Box plot (also known as Whisker plot):

<br>
This is related to the normal distribution curve and the standard deviations:

```python
salesDF.boxplot("Shipping Cost", by="Ship Mode")
#salesDF.boxplot("Shipping Cost", by="Ship Mode", figsize=(10,10))
```

From a look at the plot, we have found that there are some data points that do not seem to lie within the normal distribution. 
<br>
We will want to examine some of these anomalies or outliers more closely...


```python
salesByDeliveryTruckDF = salesDF[salesDF["Ship Mode"] == "Delivery Truck"]
```

```python
shippingCostOfSalesByDeliveryTruck = salesByDeliveryTruckDF["Shipping Cost"]
```

```python
shippingCostOfSalesByDeliveryTruck.describe()
```

```python
deliveryTruckCostStdDeviation = shippingCostOfSalesByDeliveryTruck.std()
deliveryTruckCostStdDeviation
```

```python
# let's be very sure of an outlier we want to examine... 
anomalyThreshold = shippingCostOfSalesByDeliveryTruck.mean() + 3 * deliveryTruckCostStdDeviation
anomalyThreshold
```

```python
salesByDeliveryTruckDF[salesByDeliveryTruckDF["Shipping Cost"] > anomalyThreshold]
```

# Customer Segmentation

Lets find out if we can group some customers from this dataset.
There are 3 types of grouping we want to explore:
1. Recency (last bought)
2. Frequency (return customer)
3. Total Sales (valuable customer)

```python
# Recency
salesDFGroupByCustomer = salesDF.groupby("Customer Name")
customerRecency = salesDFGroupByCustomer["Order Date"].max()
customerRecency.sort_values(ascending=False)
```

```python
# Frequency
salesDFGroupByCustomer = salesDF.groupby("Customer Name")
customerFrequency = salesDFGroupByCustomer.size()
customerFrequency.sort_values(ascending=False)
```

```python
# Amount of Sales
salesDFGroupByCustomer = salesDF.groupby("Customer Name")
customerSales = salesDFGroupByCustomer["Sales"].sum()
customerSales.sort_values(ascending=False)
```

```python
customerSales.hist(bins=30)
```

# Pareto Principle

The Pareto principle (also known as the 80/20 rule, the law of the vital few) states that, for many events, roughly 80% of the effects come from 20% of the causes. <br>
When applied to sales data it means 80% of a company's revenue is generated by 20% of its total customers





```python
netGlobalSales = customerSales.sum()
netGlobalSales
```

```python
top20CustomerSalesThreshold = customerSales.quantile(0.8)
top20CustomerSalesThreshold
```

```python
top20CustomerSales = customerSales[customerSales > customerSales.quantile(0.8)]
top20CustomerSales
```

```python
top20GlobalSales = top20CustomerSales.sum()
top20GlobalSales
```

```python
# lets check if 80/20 rule holds
top20GlobalSales/netGlobalSales
```
