# Predictive Analytics and Machine Learning

Predictive Analytics is about using all the facts at your disposal to forecast an outcome. <br>
The tool used to make such a forecast is typically called a decision model.<br>
Decision models can come about via wisdom gathered and summarized by a domain expert.<br>
However, it is also possible to create a model near to what a domain expert can come up with from just data points!</br>
Machine learning is such a technique to feed data into a model in order for it to create its own logical rules to predict an outcome.<br>

```python
import pandas
import sklearn
from sklearn_pandas.dataframe_mapper import DataFrameMapper
%matplotlib inline
```

```python
salesDF = pandas.read_excel("SampleSalesData.xlsx")
```

```python
salesDF
```

# Classification

We will start off with a simple model. <br>
Decision trees are used to classify into outcomes that are categorical. <br>
We use decision trees because it is easier to visualize how the decision of the classification came about.<br>

<img src="./images/IkBzK.png" style="height:400px"/>

Note: to generate a decision tree image in python you will need a 3rd party application called graphviz


```python
# refer to http://scikit-learn.org/stable/modules/tree.html for more documentation
from sklearn.tree import DecisionTreeClassifier
```

```python
# let's create our categorical field
salesDF["MadeProfit"]  = salesDF.Profit >= 0
```

```python
# we will create a testing and training dataset
from sklearn.model_selection import train_test_split

# random state is used to make this random state repeatable.
trainDF, testDF = train_test_split(salesDF, test_size=0.2, random_state=777)
```

```python
len(trainDF), len(testDF)
```

## Feature Engineering

One of the most time consuming parts of creating a predictive model is the creation of features to be used.

```python
# refer to http://scikit-learn.org/stable/modules/preprocessing.html on the different ways to engineer features
# these are some simple strategies used in feature engineering.
# note that there are countless ways to process data to fit into a model... refining a data model usually means refining the engineered features
from sklearn.preprocessing import LabelBinarizer
```

```python
# refer to https://github.com/scikit-learn-contrib/sklearn-pandas 

featureMapper = DataFrameMapper([
    ("Ship Mode", LabelBinarizer()),
])
```

```python
featureMapper.fit_transform(trainDF)
```

```python
featureMapper.transformed_names_
```

```python
# let's do it again but fit on the training dataset
featureMapper = DataFrameMapper([
    ("Ship Mode", LabelBinarizer()),
    ("Order Priority", LabelBinarizer())
])
```

```python
trainFeatureVectors = featureMapper.fit_transform(trainDF)
```

```python
trainFeatureVectors
```

# Creating the Model

We will be using a decision tree model since it is one of the more classical models used in machine learning. <br>
It may be superseded by other models, thus it is <b>highly recommended</b> to check out the sklearn documentation.<br>
http://scikit-learn.org/stable/



```python
dtClassifier = DecisionTreeClassifier()
```

```python
trainLabels = trainDF.MadeProfit.tolist()
```

```python
dtClassifier.fit(trainFeatureVectors, trainLabels)
```

```python
# let's visualize what the decision tree would look like
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO 
import pydotplus
from IPython.display import Image

dot_data = StringIO()

export_graphviz(dtClassifier, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True, feature_names=featureMapper.transformed_names_)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())

```

### For those having difficulty getting Graphviz integrated into your environment, you may find the a pre-exported png file sampleDecisionTree_1.png in your folder.

How would we tell if a model is good? We simply apply it to the test dataset. <br>

Do note, the test dataset must go through the same preprocessing steps as the training dataset.<br>
Also, if there are new categories or data points in the test dataset; they are not taken into consideration in the model. This is very important to note when you put models into production<br>


```python
testLabels = testDF.MadeProfit.tolist()
```

```python
# Note that we use transform here instead of fit_transform
testFeatureVectors = featureMapper.transform(testDF)
```

```python
predicted = dtClassifier.predict(testFeatureVectors)
```

We can count the difference between the testLabels (ground truth) vs the predicted labels to know how well a model performs <br>
However, there is a better way to represent the findings; in a confusion matrix. <br>




```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(testLabels, predicted))
```

```python
trainDF, testDF = train_test_split(salesDF, test_size=0.2, random_state=777)
featureMapper = DataFrameMapper([
    ("Customer Segment", LabelBinarizer()),
    ("Product Category", LabelBinarizer()),
])

trainFeatureVectors = featureMapper.fit_transform(trainDF)
testFeatureVectors = featureMapper.transform(testDF)

trainLabels = trainDF.MadeProfit.tolist()
testLabels = testDF.MadeProfit.tolist()


dtClassifier = DecisionTreeClassifier()
dtClassifier.fit(trainFeatureVectors, trainLabels)

predicted = dtClassifier.predict(testFeatureVectors)
```

```python
print(classification_report(testLabels, predicted))
```

```python
dot_data = StringIO()

export_graphviz(dtClassifier, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True, feature_names=featureMapper.transformed_names_)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
