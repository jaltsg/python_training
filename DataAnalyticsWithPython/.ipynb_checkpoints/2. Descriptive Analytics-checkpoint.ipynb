### Histograms
One of the most commonly used chart during the data exploration phase is the histogram. <br/>
The histogram allows a quick and accurate representation of the distribution of numerical data in a dataset.<br/>
The type of distribution of the chart often gives us an idea of possible data gaps, preprocessing that needs to be done.<br/>

```python
# the histogram over the larger dataset may seem like a normal distribution at first..
# however, since we already know that the shipping cost is metadata for shipping, it is natural to
# split it by category under shipping

salesDF.hist(column="Shipping Cost", by="Ship Mode")
```

```python
# the granularity can be controlled by the number of bins that we want
salesDF.hist(column="Shipping Cost", by="Ship Mode", bins=30)
```

```python
# we use the describe function again this time but to look closely at the shipping cost.
salesDF["Shipping Cost"].describe(())
```

```python
# we can group the data by a certain column before extracting some descriptive statistics off it
salesGroupedByShipMode = salesDF.groupby("Ship Mode")
```

```python
shippingCostGroupedByShipMode = salesGroupedByShipMode["Shipping Cost"]
```

```python
# note how this is similar to a pivot table
shippingCostGroupedByShipMode.describe()
```

## Anomaly Detection (Outlier Analysis)

```python
salesByDeliveryTruckDF = salesDF[salesDF["Ship Mode"] == "Delivery Truck"]
```

```python
shippingCostOfSalesByDeliveryTruck = salesByDeliveryTruckDF["Shipping Cost"]
```

```python
shippingCostOfSalesByDeliveryTruck.describe()
```

```python
deliveryTruckCostStdDeviation = shippingCostOfSalesByDeliveryTruck.std()
deliveryTruckCostStdDeviation
```

```python
# let's be very sure of an outlier we want to examine... 
anomalyThreshold = shippingCostOfSalesByDeliveryTruck.mean() + 3 * deliveryTruckCostStdDeviation
anomalyThreshold
```

```python
salesByDeliveryTruckDF[salesByDeliveryTruckDF["Shipping Cost"] > anomalyThreshold]
```
